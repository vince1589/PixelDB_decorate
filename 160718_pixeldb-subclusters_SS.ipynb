## Determine Subclusters

```python
#set parameters
similarity_threshold = .95

#make data structures
subcluster_dict = dict() #dict[receptor] = [similar receptor1, similar receptor 2, etc]

def find_largest_remaining_subcluster(similarity_table):
for receptor_parent in similarity_table.keys():
    receptor_list = list()
    for receptor_child in similarity_table.keys():
        if similarity_table.loc[receptor_child,receptor_parent] >= similarity_threshold and receptor_child != receptor_parent:
            receptor_list.append(receptor_child)
    subcluster_dict[receptor_parent] = receptor_list
subcluster_dict
```

## Add the peptides within a subclusters to protein pdb

```python
all_receptors_old = glob.glob("/scratch/users/swans/decorate_PixelDB_project/PixelDB/clusters/1/*.pdb")

receptor_directory_old = "/scratch/users/swans/decorate_PixelDB_project/PixelDB/clusters/1/"
receptor_directory_new = '/scratch/users/swans/decorate_PixelDB_project/subclusters/1/'

if not os.path.exists(receptor_directory_new):
    os.makedirs(receptor_directory_new)

for receptor_filename_old in all_receptors_old:
    receptor_id = receptor_filename_old[-16:-10]
    print(receptor_id)
    with open(receptor_directory_new+receptor_id+'_1','w') as receptor_new:
        with open(receptor_filename_old,'r') as receptor_parent:
            for line in receptor_parent:
                receptor_new.write(line)
        for receptor_child_id in subcluster_dict[receptor_id]:
            child_global_path = glob.glob(receptor_directory_old+receptor_child_id+'*'+'.pdb')[0]
            with open(child_global_path,'r') as receptor_child:
                chain_query = 'ATOM................[^'+receptor_child_id[-1:]+']'
                for line in receptor_child:
                    if re.search(chain_query,line) != None:
                        receptor_new.write(line)
```
