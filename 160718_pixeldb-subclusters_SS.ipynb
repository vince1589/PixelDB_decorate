# How number of maximal cliques vary with similarity threshold

```python
thresh_range = .5
#aint nobody fuckin w my
max_cliques = list()
avg_cliques = list()
num_cliques = list()
for threshold in np.arange(1-thresh_range,1,.01):
    print(threshold)
    simil_graph = nx.Graph(simil > threshold)
    clique_list = list(nx.find_cliques(simil_graph))
    max_cliques += [nx.graph_clique_number(simil_graph)]
    avg_cliques += [np.mean([len(x) for x in clique_list])]
    num_cliques += [len(clique_list)]
```

### .5 to 1

```python
plt.plot(np.arange(1-thresh_range,1,.01),np.array(max_cliques),label='Size of Largest Clique')
plt.plot(np.arange(1-thresh_range,1,.01),np.array(num_cliques),label='Number of Cliques')
plt.plot(np.arange(1-thresh_range,1,.01),np.array(avg_cliques),label='Avg Size of Cliques')
plt.xlabel('similarity cutoff')
plt.ylabel('count')
plt.legend(loc='upper right')
```

```python
lowerbound = 25
x_range = np.arange(1-thresh_range,1,.01)[lowerbound:]
plt.plot(x_range,np.array(max_cliques)[lowerbound:],label='Size of Largest Clique')
plt.plot(x_range,np.array(num_cliques)[lowerbound:],label='Number of Cliques')
plt.plot(x_range,np.array(avg_cliques)[lowerbound:],label='Avg Size of Cliques')
plt.xlabel('similarity cutoff')
plt.ylabel('count')
plt.legend(loc='upper right')



#conclusion: use .95 as threshold :) 
```

# Combining to Make Script

```python
import numpy as np
import pandas as pd
import glob
import os
import regex as re
import matplotlib.pyplot as plt
import subprocess
import os
import networkx as nx
%matplotlib inline

#NOTE: may need to update directories

pixelDB_cluster_path = '/scratch/users/swans/decorate_PixelDB_project/PixelDB/clusters/'
subcluster_path = '/scratch/users/swans/decorate_PixelDB_project/subclusters/'

#Get all the cluster folder numbers in a list
cluster_folders = subprocess.check_output('ls '+ pixelDB_cluster_path,shell=True)
cluster_folders = str.split(cluster_folders.decode("utf-8").rstrip(),'\n')

for cluster in cluster_folders:
    print('cluster ',cluster)
    os.system('mkdir '+ subcluster_path + cluster)
    simil = pd.read_csv('/scratch/users/swans/decorate_PixelDB_project/PixelDB/clusters/' + cluster + '/' + cluster + '_simil.CSV',delimiter=' ',index_col=0)
    cluster_graph = nx.Graph(simil >= .95)
    maximal_cliques = list()
    while cluster_graph.order() > 0:
        all_cliques = list(nx.find_cliques(cluster_graph))
        all_cliques.sort(key=len,reverse=True)
        maximal_cliques += [all_cliques[0]]
        #remove nodes from graph
        cluster_graph.remove_nodes_from(all_cliques[0])
#         print(cluster_graph.order())
#         print(len(maximal_cliques))
    print(len(maximal_cliques),' cliques identified')
    for i in range(len(maximal_cliques)):
        for j in range(len(maximal_cliques[i])):
            maximal_cliques[i][j] = maximal_cliques[i][j] + '*.pdb'
    for subcluster_folder in range(len(maximal_cliques)):
        new_dir = subcluster_path + cluster +  '/' + str(subcluster_folder+1)
        try:
            os.mkdir(new_dir)
        except OSError:
            pass
        os.system('cp '+ pixelDB_cluster_path + cluster + '/{' +','.join(maximal_cliques[subcluster_folder])+'} '+new_dir)

```
