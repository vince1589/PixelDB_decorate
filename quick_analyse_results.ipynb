```python
import pickle as pk
import glob as glob
import pandas as pd
import copy
```

```python
#Get All Pickle File
PickleFiles = glob.glob("./1A1M_decorate_results/pickle_result_files//*.pk")
print(len(PickleFiles))
```

```python
QueryName = "1A1M_A_C_1_1"
```

```python
QueryRes = dict()
BestRes = dict()

for pickfile in PickleFiles:
    
    pickname = pickfile.split("/")[-1].split(".")[0]
    
    
    
    TempRes = pk.load(open(pickfile,"rb"))
    if QueryName == pickname:
        QueryRes = copy.deepcopy(TempRes)
    print(pickfile,pickname)
    #Find Best Ali/RMSD
    for kp in TempRes.keys():
        if kp not in BestRes:
            BestRes[kp] = copy.deepcopy(TempRes[kp])
            BestRes[kp]["RecepMatch"] = pickname
        if TempRes[kp]["TotAli"] >  BestRes[kp]["TotAli"]:
            BestRes[kp] = copy.deepcopy(TempRes[kp])
            BestRes[kp]["RecepMatch"] = pickname
        if (TempRes[kp]["TotAli"] ==  BestRes[kp]["TotAli"]) & (TempRes[kp]["rmsd"] <  BestRes[kp]["rmsd"]):
            BestRes[kp] = copy.deepcopy(TempRes[kp])
            BestRes[kp]["RecepMatch"] = pickname
```

```python
DF = pd.DataFrame(BestRes).transpose()
DF["TM"] = DF["TotAli"]/DF["nres"]
```

```python
DF.sort_values(by=["TotAli","PepRMSD"],ascending=[False,True]).head(20)
```

```python

fpr, tpr, thresholds = metrics.roc_curve(DF["TM"] > 0.99,-DF["score"],  pos_label=True)
metrics.auc(fpr, tpr)
```
